```python
# https://pandas.pydata.org/pandas-docs/stable/comparison_with_r.html

import numpy as np
import pandas as pd
```

```python
# Create Pandas df
# https://cmdlinetips.com/2018/01/how-to-create-pandas-dataframe-from-multiple-lists/

test = {  'Name' : ['Jack','Jack','Jack','Jack','Jack','Jack',
                     'Al','Al','Al','Al','Al','Al','Al','Al',
                     'Sam','Sam','Sam','Sam','Sam','Sam','Sam','Sam','Sam'],
                      
          'Date' : ['2011-10-09','2011-11-09','2011-09-09','2012-10-09','2012-11-09','2013-09-09',
                     '2011-12-09','2011-11-08','2011-11-06','2011-12-29','2012-12-09','2012-11-08','2012-11-06','2012-12-29',
                     '2011-10-09','2011-10-19','2010-10-09','2010-10-09','2010-01-09','2012-10-09','2011-10-09','2013-10-09','2013-11-09'],
                      
           'Town': ['Urbana','Champaign','Urbana','Urbana','Champaign','Urbana',
                     'Dayton','Dayton','Centerville','Kettering','Dayton','Dayton','Xenia','Xenia',
                     'Lafayette','Lafayette','New Orleans','Lafayette','Lafayette','New Orleans','Lafayette','Lafayette','New Orleans'],
           
           'Sales': [2000,3000,3050,5000,6000,6050,
                     4000,6000,2000,1500,4500,6500,2500,5500,
                     2500,3000,4000,3500,3300,4300,2800,3800,4800]}


#Default: columns are ordered by name
df_test = pd.DataFrame(test)

#Order columns manually
df_test = pd.DataFrame(test,columns=['Name','Date','Town','Sales'])
```

```python
#Rename Columns(Not inplace)
df_test.rename(columns={'Town':'City'})

#Rename Columns(Inplace)
#df_test.rename(columns={'Town':'City'}, inplace = True)

#df_test
```

```python
#Select Columns
df_test.loc[:,['Name','Town']]
```

```python
#Filter by condition
df_test.loc[df_test.Town == 'Dayton',:]
```

```python
#Select Rows by Position
df_test.iloc[0:5]
```

```python
#Transform/Mutate column
df_test.assign(Sales = df_test.Sales/1000)
```

```python
#Transform/Mutate column
df_test.assign(Sales2 = df_test.Sales/1000)
```

```python
#Proportion of sales for a subgroup
df_test.assign(SubGroup_Proportion = df_test.Sales / df_test.groupby('Name')['Sales'].transform('sum'))

df_test
```

```python
#Proportion of sales for a subgroup
df_test.assign( SubGroup_Proportion = df_test.Sales / df_test.groupby('Name')['Sales'].transform('sum') )

df_test.assign( OverAll_Proportion = df_test.Sales / df_test.Sales.sum() )


df_test
```

```python
#Replace a value according to some condition
df_test.loc[(df_test.Sales == 2000) & (df_test.Town == 'Urbana'),'Sales'] = 450

df_test
```
